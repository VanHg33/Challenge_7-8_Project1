```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Read each quarter for each year of Metropolitan Median House Price Data Files and store into Pandas DataFrames

Q1_2015 = pd.read_csv(Path("Resources/2015/Metropolitan Median House Sales Q1 2015.csv"))
Q2_2015 = pd.read_csv(Path("Resources/2015/Metropolitan Median House Sales Q2 2015.csv"))
Q3_2015 = pd.read_csv(Path("Resources/2015/Metropolitan Median House Sales Q3 2015.csv"))
Q4_2015 = pd.read_csv(Path("Resources/2015/Metropolitan Median House Sales Q4 2015.csv"))

Q1_2016 = pd.read_csv(Path("Resources/2016/Metropolitan Median House Sales Q1 2016.csv"))
Q2_2016 = pd.read_csv(Path("Resources/2016/Metropolitan Median House Sales Q2 2016.csv"))
Q3_2016 = pd.read_csv(Path("Resources/2016/Metropolitan Median House Sales Q3 2016.csv"))
Q4_2016 = pd.read_csv(Path("Resources/2016/Metropolitan Median House Sales Q4 2016.csv"))

Q1_2017 = pd.read_csv(Path("Resources/2017/Metropolitan Median House Sales Q1 2017.csv"))
Q2_2017 = pd.read_csv(Path("Resources/2017/Metropolitan Median House Sales Q2 2017.csv"))
Q3_2017 = pd.read_csv(Path("Resources/2017/Metropolitan Median House Sales Q3 2017.csv"))
Q4_2017 = pd.read_csv(Path("Resources/2017/Metropolitan Median House Sales Q4 2017.csv"))

Q1_2018 = pd.read_csv(Path("Resources/2018/Metropolitan Median House Sales Q1 2018.csv"))
Q2_2018 = pd.read_csv(Path("Resources/2018/Metropolitan Median House Sales Q2 2018.csv"))
Q3_2018 = pd.read_csv(Path("Resources/2018/Metropolitan Median House Sales Q3 2018.csv"))
Q4_2018 = pd.read_csv(Path("Resources/2018/Metropolitan Median House Sales Q4 2018.csv"))

Q1_2019 = pd.read_csv(Path("Resources/2019/Metropolitan Median House Sales Q1 2019.csv"))
Q2_2019 = pd.read_csv(Path("Resources/2019/Metropolitan Median House Sales Q2 2019.csv"))
Q3_2019 = pd.read_csv(Path("Resources/2019/Metropolitan Median House Sales Q3 2019.csv"))
Q4_2019 = pd.read_csv(Path("Resources/2019/Metropolitan Median House Sales Q4 2019.csv"))

Q1_2020 = pd.read_csv(Path("Resources/2020/Metropolitan Median House Sales Q1 2020.csv"))
Q2_2020 = pd.read_csv(Path("Resources/2020/Metropolitan Median House Sales Q2 2020.csv"))
Q3_2020 = pd.read_csv(Path("Resources/2020/Metropolitan Median House Sales Q3 2020.csv"))
Q4_2020 = pd.read_csv(Path("Resources/2020/Metropolitan Median House Sales Q4 2020.csv"))

Q1_2021 = pd.read_csv(Path("Resources/2021/Metropolitan Median House Sales Q1 2021.csv"))
Q2_2021 = pd.read_csv(Path("Resources/2021/Metropolitan Median House Sales Q2 2021.csv"))
Q3_2021 = pd.read_csv(Path("Resources/2021/Metropolitan Median House Sales Q3 2021.csv"))
Q4_2021 = pd.read_csv(Path("Resources/2021/Metropolitan Median House Sales Q4 2021.csv"))

Q1_2022 = pd.read_csv(Path("Resources/2022/Metropolitan Median House Sales Q1 2022.csv"))
Q2_2022 = pd.read_csv(Path("Resources/2022/Metropolitan Median House Sales Q2 2022.csv"))
Q3_2022 = pd.read_csv(Path("Resources/2022/Metropolitan Median House Sales Q3 2022.csv"))
Q4_2022 = pd.read_csv(Path("Resources/2022/Metropolitan Median House Sales Q4 2022.csv"))

Q1_2023 = pd.read_csv(Path("Resources/2023/Metropolitan Median House Sales Q1 2023.csv"))
Q2_2023 = pd.read_csv(Path("Resources/2023/Metropolitan Median House Sales Q2 2023.csv"))

# Combine quarters into individual years
data_2015 = pd.concat([Q1_2015, Q2_2015, Q3_2015, Q4_2015], axis =1)

data_2016 = pd.concat([Q1_2016, Q2_2016, Q3_2016, Q4_2016], axis =1)

data_2017 = pd.concat([Q1_2017, Q2_2017, Q3_2017, Q4_2017], axis =1)

data_2018 = pd.concat([Q1_2018, Q2_2018, Q3_2018, Q4_2018], axis =1)

data_2019 = pd.concat([Q1_2019, Q2_2019, Q3_2019, Q4_2019], axis =1)

data_2020 = pd.concat([Q1_2020, Q2_2020, Q3_2020, Q4_2020], axis =1)

data_2021 = pd.concat([Q1_2021, Q2_2021, Q3_2021, Q4_2021], axis =1)

data_2022 = pd.concat([Q1_2022, Q2_2022, Q3_2022, Q4_2022], axis =1)

data_2023 = pd.concat([Q1_2023, Q2_2023], axis =1)
```
