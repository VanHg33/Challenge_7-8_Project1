**Data Cleaning for Year 2015**
-----------------------------

```python
# Combine the data of 2015 into a single dataset.  
data_2015= pd.concat([Q1_2015, Q2_2015, Q3_2015, Q4_2015], ignore_index=True)
# Display 
data_2015
```

```python
# Rename and drop columns: keep columns of sales and median price for each quarter of 2015
rename_df = data_2015.rename(columns ={"Sales\n1Q 2015": "Sales Q1 2015", 
                                       "Median\n1Q 2015": "Median Q1 2015", 
                                       "Sales\n2Q 2015": "Sales Q2 2015",
                                       "Median\n2Q 2015": "Median Q2 2015",
                                       "Sales\n3Q 2015": "Sales Q3 2015",
                                       "Median\n3Q 2015": "Median Q3 2015",
                                       "Sales\n4Q 2015": "Sales Q4 2015",
                                       "Median\n4Q 2015": "Median Q4 2015",
})

data_2015_reduction = rename_df[["City","Suburb", "Sales Q1 2015", "Median Q1 2015", "Sales Q2 2015",\
                                  "Median Q2 2015", "Sales Q3 2015", "Median Q3 2015", "Sales Q4 2015", "Median Q4 2015"]]

# Display 
data_2015_reduction
```

**Create new dataset and export as output data for year 2015**
------------------------------
As part of data exploration and cleaning steps, new dataset has been created for future data visualisation. 

```python
# Check data type
data_2015_reduction.dtypes
```

```python
# Get the number of sale before group by City & Suburb 
sale_total_cols = ["Sales Q1 2015","Sales Q2 2015", "Sales Q3 2015",  "Sales Q4 2015"]
median_total_cols= ["Median Q1 2015", "Median Q2 2015","Median Q3 2015", "Median Q4 2015"]
sale_2015 = data_2015_reduction.assign(sale=data_2015_reduction[sale_total_cols].sum(1)).drop(median_total_cols, 1)
sale_2015["2015 Sales"]=sale_2015["sale"]
sale_2015
```

```python
# Group data by columns of "City" and "Suburb" and calculate the mean of number of sale of each suburb of year 2015
groupby_sale_2015 = round(sale_2015.groupby(["City","Suburb"], sort=False).mean(),2)
groupby_sale_2015
# Display DataFrame
sale_2015_df = pd.DataFrame(groupby_sale_2015.drop(groupby_sale_2015.columns[[4]], axis=1))
sale_2015_df
```

```python
# Export csv file for the number of sale of house by city and suburb in year 2015
```

```python
# Get the median price of sale before group by City & Suburb 
median_2015 = data_2015_reduction.assign(median=data_2015_reduction[median_total_cols].sum(axis=1)).drop(sale_total_cols, 1)
median_2015["2015 Median Price"]=median_2015["median"]
median_2015
```

```python
# Group data by columns of "City" and "Suburb" and calculate the mean of median prices of each suburb of year 2015
groupby_median_2015 = median_2015.groupby(["City","Suburb"], sort=False).median()
# Display DataFrame and drop null column
median_2015_df = pd.DataFrame(groupby_median_2015.drop(groupby_median_2015.columns[[4]], axis=1))
median_2015_df
```

```python
# Combine 2 DataFrames and drop row with no sale
new_df_2015 = pd.concat([sale_2015_df, median_2015_df], axis=1 )
drop_zero_combine_df = new_df_2015[new_df_2015["2015 Sales"] != 0]
drop_zero_combine_df
```

```python
# Export csv file for the median sale prices by city and suburb in year 2015
```

**Create Dataset only contain the Mean House Price and the Mean of Sale for year 2015**
-------------------------

```python
mean_sale_2015_df = pd.DataFrame(groupby_sale_2015.drop(groupby_sale_2015.columns[[0,1,2,3,4]], axis=1))
median_price_2015_df = pd.DataFrame(groupby_median_2015.drop(groupby_median_2015.columns[[0,1,2,3,4]], axis=1))
endpoint_df = pd.concat([mean_sale_2015_df,median_price_2015_df], axis=1)
df_2015 = endpoint_df[endpoint_df["2015 Sales"] != 0]
df_2015 
```

```python
# Export csv file 
df_2015.to_csv("Output_data/2015_combine.csv", index_label="City")
```
